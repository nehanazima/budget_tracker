## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
[Help me write a program that allows users to input their monthly income, allows users to input up to 3 expenses, calculates remaining budget,shows a warning if budget is negative]
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# Ask for income
income  = float(input("Enter your monthly income: "))

# Ask for expenses
expense1 = float(input("Enter expense 1: "))
expense2 = float(input("Enter expense 2: "))
expense3 = float(input("Enter expense 3: "))

# Calculate total expenses
total_expenses = expense1 + expense2 + expense3

# Calculate remaining budget
remaining = income - total_expenses

# Show results
print("Total expenses:", total_expenses)
print("Remaining budget:", remaining)

#Warning if budget is negative
if remaining < 0:
  print("Warning: You are over your budget!")
else:
  print("Good job! You are within your budget.")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: Input validation (no negative numbers)"
enhancement_prompt = """
[Add input validation to my budget tracker so that the user cannot enter negative numbers for income or expenses. If a negative number is entered, display an error message and ask the user to enter the value again.]
"""

print(f"🚀 ENHANCEMENT: {enhancement_choice}")
```

```python
# ENHANCED BUDGET TRACKER
def get_positive_number(prompt):
  while True:
    try:
      value = float(input(prompt))
      if value < 0:
        print("Error: Please enter a positive number.")
      else:
        return value
    except ValueError:
      print("Invalid input. Please enter a number.")

print("Welcome to Your Budget Tracker")

#Get income
income = get_positive_number("Enter your monthly income: ")

#Get expenses
rent = get_positive_number("Enter rent amount: ")
food = get_positive_number("Enter food expense: ")
transport = get_positive_number("Enter transport expense: ")

#Calculate total expenses
total_expenses = rent + food + transport

#Calculate remaining balance
balance = income - total_expenses

print("\n Budget Summary")
print(f"Total Income: ${income}") # Corrected to display income, not total_expenses
print(f"Total Expenses: ${total_expenses}") # Added to show total expenses
print(f"Remaining Balance: ${balance}")

#Budget check
if balance > 0:
  print("Good job! You are within your budget")
elif balance == 0:
  print("You have used all your income.")
else:
  print("You are over budget!")
```
